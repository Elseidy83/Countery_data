```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
import seaborn as sns

sns.set(rc={'figure.figsize': [7, 7]}, font_scale=1.2)
```

```python
df = pd.read_csv('Country-data.csv')
df
```

```python
ds = df.drop(['country'],axis=1)
```

```python
sns.heatmap(ds.corr(), annot=True, fmt='.1f')
```

```python
def get_sum(rw):
    return rw['child_mort']+ rw['exports']+rw['health']+rw['imports']+rw['income']+rw['inflation']+rw['life_expec']+rw['total_fer']+rw['gdpp']
```

```python
dd = ds.corr().abs()
```

```python
dd.apply(get_sum).sort_values(ascending=False)
```

```python
ds = ds.drop(['inflation','imports','health'],axis=1)
```

```python
ds
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(ds)
```

```python
x_scaled
```

```python
sns.pairplot(ds)
```

```python
plt.scatter(x_scaled[:, 0], x_scaled[:, 1])
```

```python
import scipy.cluster.hierarchy as sch
```

```python
dendrogram = sch.dendrogram(sch.linkage(ds, method='ward'))
```

```python
model = AgglomerativeClustering(n_clusters=5)
clusters = model.fit_predict(x_scaled)
clusters
```

```python
plt.scatter(x_scaled[:, 0], x_scaled[:,7], c=clusters, cmap='viridis')
```

```python
df['Clusters'] = clusters
df
```

```python
df.groupby('Clusters').describe().transpose()
```
